```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
df =pd.read_csv("FuelConsumption.csv")
df.head(11)
```

```python
df = df[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_CITY','FUELCONSUMPTION_HWY','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
```

```python
df.head(20)
```

```python
plt.scatter(df.CYLINDERS, df.CO2EMISSIONS, color = 'green', alpha = 0.1)
plt.xlabel("CYLINDERS")
plt.ylabel("CO2EMISSIONS")
plt.show()
```

```python
x = df['CYLINDERS']
y = df['CO2EMISSIONS']
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.4, random_state = 23)
```

```python
train_x
```

```python
train_x = np.array(train_x).reshape(-1,1)
test_x = np.array(test_x).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(train_x, train_y)
```

```python
c = reg.intercept_
m = reg.coef_
c,m
```

```python
predtrain_y = reg.predict(train_x)
```

```python
plt.scatter(df.CYLINDERS, df.CO2EMISSIONS, color = 'green', alpha = 0.1)
plt.xlabel("CYLINDERS")
plt.ylabel("CO2EMISSIONS")
plt.plot(train_x, predtrain_y, color = 'red')
plt.show()
```

```python
predtest_y = reg.predict(test_x)
```

```python
plt.scatter(df.CYLINDERS, df.CO2EMISSIONS, color = 'green', alpha = 0.1)
plt.xlabel("CYLINDERS")
plt.ylabel("CO2EMISSIONS")
plt.plot(test_x, predtest_y, color = 'red')
plt.show()
```
